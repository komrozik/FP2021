```python
import numpy as np
import matplotlib
font = {'size': 11.0}
matplotlib.rc('font', **font)
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
import uncertainties
from uncertainties import ufloat
import uncertainties.unumpy as unp 
from uncertainties.unumpy import (nominal_values as noms, std_devs as stds)
```

```python
df1 = pd.read_csv('data/data1'
                 ,sep = ','
                 ,lineterminator='\n'
                 ,skiprows=(2)
                 ,header = (0)
                 )
df2 = pd.read_csv('data/data2'
                 ,sep = ','
                 ,lineterminator='\n'
                 ,header = (0)
                 ,skiprows = (2)
                 )
```

```python
df1['A'][0:12] = -df1['A'][0:12]
```

```python
df1 = df1.drop(4)
```

```python
df2 = df2.drop(1)
```

```python
def A2(tau,a,b):
    return a*(1-2*np.exp(-(tau/b)))

params1,cov = curve_fit(A2,df1['tau'],df1['A'])
errors = np.sqrt(np.diag(cov))
params1_err = unp.uarray(params1,errors)

plt.figure(figsize=(6.4,3.96),dpi=300)
plt.plot(df1['tau']
         ,df1['A']
         ,ls = 'None'
         ,marker = 'x'
         ,label = 'Messdaten'
        )
plt.plot(df1['tau']
         ,A2(df1['tau'],*params1)
         ,alpha = 0.5
         ,label = 'Fit'
        )
plt.xscale('log')
plt.legend(loc = 'best')
plt.tight_layout()
plt.savefig('plots/FID_amp_1.pdf',bbox_inches = "tight")
None
```

```python
T_2 = 0.0110

def A(tau,a,b,c):
    return a*np.exp(-(2*tau)/(c))*np.exp(-(tau**3)/(b))

params2,cov = curve_fit(A,df2['tau'],df2['A'])
errors = np.sqrt(np.diag(cov))
params2_err = unp.uarray(params2,errors)


plt.figure(figsize=(6.4,3.96),dpi=300)
plt.plot(df2['tau']
         ,df2['A']
         ,ls = 'None'
         ,marker = 'x'
         ,label = 'Messdaten'
        )
plt.plot(df2['tau']
         ,A(df2['tau'],*params2)
         ,alpha = 0.5
         ,label = 'Fit'
        )
plt.plot(np.linspace(0,0.05,1000)
         ,A(np.linspace(0,0.05,1000),*params2)
         ,alpha = 0.5
         ,label = 'Fit linspace'
        )
plt.xscale('log')
plt.legend(loc = 'best')
```

```python
def f(A,tau):
    return np.log(A)-2*(tau)/T_2
```

```python
plt.figure(figsize=(6.4,3.96),dpi=300)
plt.plot(df2['tau']**3
         ,f(df2['A'],df2['tau'])
         ,ls = 'None'
         ,marker = 'x'
         ,label = 'Messdaten'
        )
plt.legend(loc = 'best')
```
